### Correct code

```python

url = 'https://www.magicbricks.com/property-for-rent/residential-real-estate?bedroom=&proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment,Service-Apartment&Locality=Wakad&cityName=Pune'

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

cards = soup.find_all('div', class_='mb-srp__card')

with open('clean_real_estate_data.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Title', 'Price_Rs', 'Furnishing', 'Bathrooms', 'Carpet_Area_sqft', 'Floor_No'])

    for card in cards:
        title = card.find('h2', class_='mb-srp__card--title')
        price = card.find('div', class_='mb-srp__card__price--amount')
        summary_items = card.find_all('div', class_='mb-srp__card__summary__list--item')

        furnish, bathroom, carpet_area, floor = '', '', '', ''

        for item in summary_items:
            data_summary = item.get('data-summary', '').strip()
            text_value = item.text.strip()

            if data_summary == 'furnishing':
                # extract actual Furnishing types only
                match = re.search(r'(Furnished|Semi-Furnished|Unfurnished)', text_value, re.IGNORECASE)
                furnish = match.group(1) if match else ''
            elif data_summary == 'bathroom':
                bathroom = re.sub(r'[^0-9]', '', text_value)
            elif data_summary == 'carpet-area':
                carpet_area = re.sub(r'[^0-9.]', '', text_value)
            elif data_summary == 'floor':
                match = re.search(r'(\d+)', text_value)
                floor = match.group(1) if match else ''

        # Clean title and price
        title = title.text.strip() if title else ''
        price = price.text.strip() if price else ''
        price = re.sub(r'[^0-9.]', '', price)

        writer.writerow([title, price, furnish, bathroom, carpet_area, floor])

print("✅ Clean data stored in 'clean_real_estate_data.csv' — ready for EDA.")
```

## Here seperated the columns from title

```python
import pandas as pd
import numpy as np
import re

df1 = pd.read_csv("clean_real_estate_multi_location1.csv")

# Clean title
df1['Title'] = df1['Title'].astype(str).str.replace(r'\s+', ' ', regex=True).str.strip()

# Extract fields
df1['BHK'] = df1['Title'].str.extract(r'(\d+)\s*(?:BHK|RK)', expand=False)
df1['BHK'] = pd.to_numeric(df1['BHK'], errors='coerce').astype('Int64')

df1['Property_Type'] = df1['Title'].str.extract(r'(?:\d+\s*(?:BHK|RK)\s*)(.*?)\s*for\b', expand=False)
df1['Listing_Type']  = df1['Title'].str.extract(r'\bfor\s+([A-Za-z]+)', expand=False)

soc_loc_city = df1['Title'].str.extract(r'\bin\s*(.+?),\s*([^,]+),\s*([^,]+)\s*$', expand=True)
df1['Society']  = soc_loc_city[0]
df1['Locality'] = soc_loc_city[1]
df1['City']     = soc_loc_city[2]

# Fix rows where society not found (only 2 parts)
mask_missing_soc = df1['Society'].isna()
two_parts = df1.loc[mask_missing_soc, 'Title'].str.extract(r'\bin\s*([^,]+),\s*([^,]+)\s*$')
df1.loc[mask_missing_soc, 'Locality'] = two_parts[0]
df1.loc[mask_missing_soc, 'City']     = two_parts[1]

# Replace string 'nan' with NaN properly
df1 = df1.apply(lambda s: s.where(s.ne('nan'), np.nan))

df1.to_csv("clean_real_estate_multi_location_parsed.csv", index=False, encoding="utf-8")
print("✅ Done! BHK cleaned and saved.")
```

### Extracted the columns of Title in nashik

```python

# 1) Load Nashik scraped csv
df = pd.read_csv("clean_real_estate_nashik.csv")

# 2) Clean title
df['Title'] = df['Title'].astype(str).str.replace(r'\s+', ' ', regex=True).str.strip()

# 3) Extract BHK number only
df['BHK'] = df['Title'].str.extract(r'(\d+)\s*(?:BHK|RK)', expand=False).astype('Int64')

# 4) Extract Property type (Flat / Studio Apartment etc.)
df['Property_Type'] = df['Title'].str.extract(r'(?:\d+\s*(?:BHK|RK)\s*)(.*?)\s*for', expand=False)

# 5) Extract listing type (Rent or Sale)
df['Listing_Type']  = df['Title'].str.extract(r'\bfor\s+([A-Za-z]+)', expand=False)

# 6) Extract everything after "in "
df['location_part'] = df['Title'].str.extract(r'\bin\s*(.*)', expand=False)

# 7) City = last part
df['City'] = df['location_part'].str.split(',').str[-1].str.strip()

# 8) Locality = second last part
df['Locality'] = df['location_part'].str.split(',').str[-2].str.strip()

# 9) Society = everything before last two comma parts
df['Society'] = df['location_part'].apply(
    lambda x: ','.join(x.split(',')[:-2]).strip() if isinstance(x, str) and x.count(',') >= 2 else np.nan
)

# 10) Convert numeric columns
df['Price_Rs'] = pd.to_numeric(df['Price_Rs'], errors='coerce')
df['Bathrooms'] = pd.to_numeric(df['Bathrooms'], errors='coerce').astype('Int64')
df['Carpet_Area_sqft'] = pd.to_numeric(df['Carpet_Area_sqft'], errors='coerce').astype('Int64')
df['Floor_No'] = pd.to_numeric(df['Floor_No'], errors='coerce').astype('Int64')

# 11) Remove helper column
df = df.drop(columns=['location_part'])

# 12) Save
df.to_csv("clean_real_estate_nashik_parsed.csv", index=False)
print("✅ Nashik cleaned & extracted successfully!")
```

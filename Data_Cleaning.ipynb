```python
df.columns = [c.strip().replace(' ', '_') for c in df.columns]
df.columns
```

```python
# fill missing Furnishing with 'Unknown'
df['Furnishing'] = df['Furnishing'].fillna('Unknown')

# drop rows where Price is missing
df = df.dropna(subset=['Price_Rs'])
```

```python
df = df.dropna(subset=['Locality','Carpet_Area_sqft','Floor_No'])
```

```python
# Fill BHK from title
df['BHK'] = df['BHK'].fillna(df['Title'].str.extract(r'(\d+)\s*BHK')[0])

# Fill Property Type from title
df['Property_Type'] = df['Property_Type'].fillna(df['Title'].str.extract(r'(Studio Apartment|1 RK|Apartment|Flat|Bungalow)', expand=False))
```

```python
df[['BHK','Property_Type']].isna().sum()
```

```python
df['BHK'] = df['BHK'].fillna(df['BHK'].mode()[0])
```

```python
df[['BHK','Property_Type']].isna().sum()
```

```python
df = df.drop(columns=['Society','Property_Type','Listing_Type','Title'])
```

```python
df.dtypes
```

```python
df['Carpet_Area_sqft'] = df['Carpet_Area_sqft'].astype(int)
df['Floor_No'] = df['Floor_No'].astype(int)
df['BHK'] = df['BHK'].astype(int)
df['Bathrooms'] = df['Bathrooms'].astype(int)
```

```python
df.dtypes
```

```python
def convert_price(x):
    x = float(x)
    if x < 1000:  
        return x * 100000  # convert lakh to rupees
    else:
        return x           # already in rupees

df['Price_Rs'] = df['Price_Rs'].apply(convert_price).astype(int)
```

```python
df['Price_per_sqft'] = df['Price_Rs'] / df['Carpet_Area_sqft']
df['Price_per_sqft'] = df['Price_per_sqft'].round(2)
```

```python
# Arangement of columns 
df = df[['Locality','City','BHK','Price_Rs','Price_per_sqft','Carpet_Area_sqft','Floor_No','Bathrooms','Furnishing']]
```

```python
# Changing the column name
df = df.rename(columns={'Price_Rs':'Rent'})
df = df.rename(columns={'Price_per_sqft':'Rent_per_sqft'})

```

```python
df.to_csv('pune_nashik_combined2.csv', index=False)
```

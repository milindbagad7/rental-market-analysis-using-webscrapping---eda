```python
def visualize_categorical(df):
    cat_df = df.select_dtypes(include='object')  # select cat cols
    
    for col in cat_df.columns:
        print(f"Column: {col}")
        print(cat_df[col].value_counts())
        print("*"*50)
        
        plt.figure(figsize=(20,5))
        
        # Barplot
        plt.subplot(1,2,1)
        a = sns.countplot(x=cat_df[col], order=cat _df[col].value_counts().index)
        for i in a.containers:
            a.bar_labels=(i)
    
        plt.title(f"Countplot of {col}")
        plt.xticks(rotation=45)
        
        # Pie chart
        plt.subplot(1,2,2)
        cat_df[col].value_counts().plot(kind='pie', autopct='%1.1f%%')
        plt.title(f"Pie Chart of {col}")
        
        plt.tight_layout()
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSV
df = pd.read_csv('pune_nashik_combined2.csv')

# --- Pune Data ---
pune = df[df['City'].str.lower() == 'pune'].copy()
pune['BHK'] = pd.to_numeric(pune['BHK'], errors='coerce')
pune['Rent'] = pd.to_numeric(pune['Rent'], errors='coerce')
avg_pune = pune.groupby('BHK')['Rent'].mean()

# --- Nashik Data ---
nashik = df[df['City'].str.lower() == 'nashik'].copy()
nashik['BHK'] = pd.to_numeric(nashik['BHK'], errors='coerce')
nashik['Rent'] = pd.to_numeric(nashik['Rent'], errors='coerce')
avg_nashik = nashik.groupby('BHK')['Rent'].mean()

# Align index (some BHKs may be missing in one city)
bhk_values = sorted(list(set(avg_pune.index) | set(avg_nashik.index)))

pune_rents = [avg_pune.get(bhk, 0) for bhk in bhk_values]
nashik_rents = [avg_nashik.get(bhk, 0) for bhk in bhk_values]

# Plot grouped bar chart
x = np.arange(len(bhk_values))
width = 0.35

plt.figure(figsize=(10,6))
plt.bar(x - width/2, pune_rents, width, label='Pune')
plt.bar(x + width/2, nashik_rents, width, label='Nashik')

plt.xlabel("BHK Type")
plt.ylabel("Average Rent (â‚¹)")
plt.title("Average Rent by BHK â€“ Pune vs Nashik")
plt.xticks(x, bhk_values)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSV
df = pd.read_csv('pune_nashik_combined2.csv')

# --- Pune Data ---
pune = df[df['City'].str.lower() == 'pune'].copy()
pune['BHK'] = pd.to_numeric(pune['BHK'], errors='coerce')
pune['Rent'] = pd.to_numeric(pune['Rent'], errors='coerce')
avg_pune = pune.groupby('BHK')['Rent'].mean()

# --- Nashik Data ---
nashik = df[df['City'].str.lower() == 'nashik'].copy()
nashik['BHK'] = pd.to_numeric(nashik['BHK'], errors='coerce')
nashik['Rent'] = pd.to_numeric(nashik['Rent'], errors='coerce')
avg_nashik = nashik.groupby('BHK')['Rent'].mean()

# Align index (some BHKs may be missing in one city)
bhk_values = sorted(list(set(avg_pune.index) | set(avg_nashik.index)))

pune_rents = [avg_pune.get(bhk, 0) for bhk in bhk_values]
nashik_rents = [avg_nashik.get(bhk, 0) for bhk in bhk_values]

# Plot grouped bar chart
x = np.arange(len(bhk_values))
width = 0.35

plt.figure(figsize=(10,6))
bars_pune = plt.bar(x - width/2, pune_rents, width, label='Pune', color='steelblue')
bars_nashik = plt.bar(x + width/2, nashik_rents, width, label='Nashik', color='seagreen')

# --- Add Bar Labels ---
for bar in bars_pune:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f"{int(height)}", 
             ha='center', va='bottom', fontsize=9)

for bar in bars_nashik:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f"{int(height)}", 
             ha='center', va='bottom', fontsize=9)

plt.xlabel("BHK Type")
plt.ylabel("Average Rent (â‚¹)")
plt.title("Average Rent by BHK â€“ Pune vs Nashik")
plt.xticks(x, bhk_values)
plt.legend()
plt.tight_layout()
plt.savefig("Avg_rent_bhk_P&N.png", dpi=300, bbox_inches='tight')

plt.show()
```
